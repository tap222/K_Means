```python
df = pd.read_excel('kmeans1.xlsx')
df.head()
```

```python
df1 = df.drop(["ID Tag", "Model", "Department"], axis = 1)
df1.head()
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=4, init='k-means++', n_init=10)
```

```python
km.fit(df1)
```

```python
x = km.fit_predict(df1)
x
```

```python
df["Cluster"]= x
df.head()
```

```python
df1 = df.sort_values(['Cluster'])
df1
```
